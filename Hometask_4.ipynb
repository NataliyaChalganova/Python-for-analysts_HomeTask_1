#### 1.2 Изучите стоимости недвижимости

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(15, 4))

sns.histplot(df['price'], bins = 500)

plt.title('Стоимость домов')
plt.xlabel('Цена')
plt.ylabel('Количество');
```

Стоимость недвижимости варьируется в довольно широких пределах, начиная от примерно 50 000 и заканчивая около 400 000, также есть дома и с более высокой стоимостью.

#### 1.3 Изучите распределение квадратуры жилой

```python
plt.figure(figsize=(15, 4))
plt.hist (df['sqft_living'], bins = 50)

#sns.histplot(df['sqft_living'], bins = 200)

plt.title('Распределение жилой площади домов')
plt.xlabel('Жилая площадь')
plt.ylabel('Количество');
```

Жилая площадь домов также очень сильно варьируется, но основная часть недвижимости имеет около 2000 квадратных метров жилой площади

#### 2.1 Изучите распределение домов от наличия вида на набережную

```python
data = df['waterfront'].value_counts(normalize = True)
data.head()
```

```python
plt.figure(figsize = (10, 4))
plt.pie (data, autopct = '%.2f%%', labels = data.index)
plt.title ('Распределение домов от наличия вида на набережную');
```

Менее одного процента домов имеют вид на набережную.

#### 2.2 Изучите распределение этажей домов

```python
data1 = df['floors'].value_counts(normalize = True)
data1.head()
```

```python
plt.figure(figsize = (4, 6))
plt.pie (data1, autopct = '%.1f%%', labels = data1.index)
plt.title ('Распределение этажей домов');
```

#### 2.3 Изучите распределение состояния домов

```python
plt.figure(figsize=(15, 4))
plt.hist (df['grade'], bins = 26)
plt.title('Распределение состояния домов')
plt.xlabel('Оценка состояния домовь')
plt.ylabel('Количество');
```

Основная масса домов имеет среднюю оценку состояния.

#### 2.4 Изучите распределение года постройки

```python
plt.figure(figsize=(15, 4))

sns.histplot(df['yr_built'], bins = 50)

plt.title('Распределение года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Количество');
```
